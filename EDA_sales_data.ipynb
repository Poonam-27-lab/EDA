```python
# Drop non-essential columns
columns_to_drop = ['ADDRESSLINE1', 'ADDRESSLINE2', 'POSTALCODE', 'PHONE', 'CONTACTFIRSTNAME', 'CONTACTLASTNAME']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
print(f"Dropped non-essential columns: {columns_to_drop}")
```

```python
df.isnull().sum()
```

```python
df[["STATE","TERRITORY"]]=df[["STATE","TERRITORY"]].fillna("Unknown")
```

```python
# Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
print("ORDERDATE converted to datetime objects.")

# Extract time-based features
df['ORDER_YEAR'] = df['ORDERDATE'].dt.year
df['ORDER_MONTH'] = df['ORDERDATE'].dt.month
df['ORDER_DAYOFWEEK'] = df['ORDERDATE'].dt.day_name()
df['ORDER_QTR'] = df['ORDERDATE'].dt.to_period('Q')
print("Extracted YEAR, MONTH, DAYOFWEEK, and QTR features.")
```

```python
df.columns
```

```python
#Univariate Analysis: Sales Distribution
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['SALES'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Sales')
plt.xlabel('Sales (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Outlier Analysis
numerical_cols = ['QUANTITYORDERED', 'PRICEEACH', 'MSRP', 'SALES']
plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
#Categorical Analysis: Product Line Orders
plt.figure(figsize=(12, 6))
sns.countplot(y='PRODUCTLINE', data=df, order=df['PRODUCTLINE'].value_counts().index, palette='viridis')
plt.title('Order Count by Product Line')
plt.xlabel('Number of Orders')
plt.ylabel('Product Line')
plt.show()
```

```python
#Categorical Analysis: Deal Size Distribution
plt.figure(figsize=(8, 6))
df['DEALSIZE'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2')
)
plt.title('Distribution of Deal Sizes')
plt.ylabel('')
plt.show()
```

```python
#Bivariate Analysis: Total Sales by Product Line
sales_by_product = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values, palette='magma')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Time Series Analysis: Monthly Sales Trend

monthly_sales = df.groupby('MONTH_ID')['SALES'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Monthly Sales Trend (All Years Combined)')
plt.xlabel('Month')
plt.ylabel('Total Sales (USD)')
plt.xticks(
    ticks=range(1, 13), 
    labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
plt.show()
```

```python
#Correlation Analysis: Numerical Features
numerical_df = df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP', 'YEAR_ID', 'MONTH_ID']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Top 10 Countries by Sales
top_10_countries = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette='plasma')
plt.title('Top 10 Countries by Total Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Order Status Analysis
status_sales = df.groupby('STATUS')['SALES'].sum().sort_values(ascending=False)
status_counts = df['STATUS'].value_counts()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(y=status_sales.index, x=status_sales.values, palette='viridis')
plt.title('Total Sales Value by Order Status')
plt.xlabel('Total Sales (USD)')

plt.subplot(1, 2, 2)
sns.barplot(y=status_counts.index, x=status_counts.values, palette='cividis')
plt.title('Order Count by Order Status')
plt.xlabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Customers by Sales
top_10_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=top_10_customers.index, x=top_10_customers.values, palette='rocket')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Total Sales (USD)')
plt.ylabel('Customer Name')
plt.show()
```
